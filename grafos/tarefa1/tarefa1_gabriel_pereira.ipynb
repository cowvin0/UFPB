```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

data = pd.read_csv("diabetes_dataset.csv")

X = data.drop(columns=["Outcome"])
y = data["Outcome"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

tree_overfit = DecisionTreeClassifier(max_depth=20, random_state=42)
tree_overfit.fit(X_train, y_train)

y_train_pred = tree_overfit.predict(X_train)
y_test_pred = tree_overfit.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Acurácia no conjunto de treino: {train_accuracy * 100:.2f}%")
print(f"Acurácia no conjunto de teste: {test_accuracy * 100:.2f}%")

plt.figure(figsize=(10, 6))
plt.plot([1, 2], [train_accuracy, test_accuracy], marker='o', linestyle='-', color='r')
plt.title("Overfitting com Árvore de Decisão (Diabetes)")
plt.xlabel("Conjunto de Dados")
plt.ylabel("Acurácia")
plt.xticks([1, 2], ['Treinamento', 'Teste'])
plt.grid(True)
plt.show()
```

Os resultados indicam um claro caso de sobreajuste. A árvore de decisão alcançou 100% de acurácia no conjunto de treino, sugerindo que memorizou completamente os dados de treinamento. No entanto, a acurácia caiu para 70,13% no conjunto de teste, o que demonstra que o modelo não generaliza bem para novos dados.
